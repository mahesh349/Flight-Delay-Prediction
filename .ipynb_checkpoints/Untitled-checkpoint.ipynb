```python
import pandas as pd
import random

# Load the CSV file into a pandas DataFrame
input_csv_path = './2018.csv'  # Replace with the path to your CSV file
output_csv_path = './output_dataset.csv'  # Replace with the desired output path

df = pd.read_csv(input_csv_path)

# Calculate the number of rows to keep (30% of the original)
num_rows_to_keep = int(0.05 * len(df))

# Get a random sample of rows to keep
rows_to_keep = random.sample(range(len(df)), num_rows_to_keep)

# Create a new DataFrame with only the selected rows
df_filtered = df.iloc[rows_to_keep]

# Save the filtered DataFrame to a new CSV file
df_filtered.to_csv(output_csv_path, index=False)

print(f"{num_rows_to_keep} rows kept, {len(df) - num_rows_to_keep} rows deleted.")
```
