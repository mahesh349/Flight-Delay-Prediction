```python
#Kushal Trivedi 20019094

import pandas as pd
import matplotlib.pyplot as plotLib

#1
csvPath = 'breast-cancer-wisconsin.csv'
readFile = pd.read_csv(csvPath)
column_stats = readFile.describe().loc[['min', 'max', 'mean']]

print("Statistics for each column:")
print(column_stats)

#2
readFile.replace('?', pd.NA, inplace=True)
rows_with_missing_values = readFile[readFile.isnull().any(axis=1)]
print("Rows with missing values:")
print(rows_with_missing_values)


#3
readFile = readFile.apply(pd.to_numeric, errors='coerce')
column_means = readFile.mean()
readFile.fillna(column_means, inplace=True)

print("DataFrame after replacing missing values:")
print(readFile)

#4
frequency_table = pd.crosstab(readFile['Class'], readFile['F6'], margins=True, margins_name='Total')

print("Frequency Table of 'Class' vs. 'F6':")
print(frequency_table)


#5
columns_to_plot = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6']
for i in range(len(columns_to_plot)):
    for j in range(i + 1, len(columns_to_plot)):
        col1 = columns_to_plot[i]
        col2 = columns_to_plot[j]
        
        plotLib.scatter(readFile[col1], readFile[col2])
        plotLib.xlabel(col1)
        plotLib.ylabel(col2)
        plotLib.title(f'Scatter Plot of {col1} vs {col2}')
        
plotLib.show()

#
columns_to_plot = ['F7', 'F8', 'F9']

#6
readFile[columns_to_plot].hist(bins=10, figsize=(12, 6))
plotLib.suptitle('Histograms for Columns F7 to F9')
plotLib.show()

readFile[columns_to_plot].boxplot(figsize=(10, 6))
plotLib.title('Box Plots for Columns F7 to F9')
plotLib.show()


#7
csvPath = 'breast-cancer-wisconsin.csv'
readFile = pd.read_csv(csvPath)

readFile = readFile[~readFile.eq('?').any(axis=1)]
print("Updated Dataset:")
print(readFile)
```
